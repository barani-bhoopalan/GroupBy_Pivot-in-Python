```python
import pandas as pd
```

```python
##Read the transaction data from Excel
InvoiceData = pd.read_excel(r'C:\Users\baran\OneDrive\Documents\Barani Interview question\Github\BB_Sales_Complete.xlsx')
```

```python
##Group by Invoice number and create dictionary groups based on it
invoice_Items_Dict = InvoiceData.groupby('InvoiceNo')['Category'].apply(lambda x: x.tolist()).to_dict()

##Convert dictionary to dataframe and align based on Index(Horizontal orientation)
invoice_Df = pd.DataFrame.from_dict(invoice_Items_Dict, orient='index')

##Dictionary values are of different length - this results in few values holding "None".
##Replace None with empty string.
invoice_Df.replace(to_replace=[None], value='', inplace=True)
invoice_Df
```

```python
##Write the dataframe itemvalues to CSV stored in local directory
invoice_Df.to_csv(r'C:\Users\baran\OneDrive\Desktop\Sample.csv', index= True)
```
